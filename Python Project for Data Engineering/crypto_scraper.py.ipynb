```python
import numpy as np
import pandas as pd
import requests
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')
```

```python
url="https://tradingeconomics.com/crypto"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'
}
response = requests.get(url, headers=headers)
tables = pd.read_html(response.text)
print(len(tables))
print(tables[0].head())
```

```python
df = tables[0]
df.columns = range(df.shape[1])
```

```python
df = df[[0,1,3]]
```

```python
df = df.iloc[0:6,:]
```

```python
df.columns = ['Crypto', 'Price', 'Daily Change (%)']
```

```python
df['Daily Change (%)'] = df['Daily Change (%)'].str.rstrip('%').astype(float)
df[['Daily Change (%)']] = np.round(df[['Daily Change (%)']], 2)
```

```python
df.rename(columns = {'Daily Change (%)': '1D Change (%)'})
```
